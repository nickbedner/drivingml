# Model Analysis (Google Colab)
This notebook loads your trained PyTorch model and visualizes its behavior.

```python
import torch
import torch.nn as nn
import numpy as np
import matplotlib.pyplot as plt
```

```python
model = nn.Sequential(
    nn.Linear(2, 64),
    nn.ReLU(),
    nn.Linear(64, 2),
    nn.Tanh()
)

model.load_state_dict(torch.load('latest_model.pt', map_location=torch.device('cpu')))
model.eval()

print('Model loaded successfully!')
```

## Inspect Model Parameters

```python
for name, param in model.named_parameters():
    print(f"{name}: mean={param.data.mean():.4f}, std={param.data.std():.4f}")
```

## Visualize Policy Output Over State Space

```python
x_vals = np.linspace(-1, 1, 50)
y_vals = np.linspace(-1, 1, 50)

actions_x = []
actions_y = []

for x in x_vals:
    row_x = []
    row_y = []
    for y in y_vals:
        state = torch.tensor([x, y], dtype=torch.float32)
        with torch.no_grad():
            action = model(state)
        row_x.append(action[0].item())
        row_y.append(action[1].item())
    actions_x.append(row_x)
    actions_y.append(row_y)

actions_x = np.array(actions_x)
actions_y = np.array(actions_y)
```

## Plot Action 1

```python
plt.figure()
plt.title('Action 1 Output')
plt.imshow(actions_x, extent=[-1,1,-1,1], origin='lower')
plt.colorbar()
plt.xlabel('Y')
plt.ylabel('X')
plt.show()
```

## Plot Action 2

```python
plt.figure()
plt.title('Action 2 Output')
plt.imshow(actions_y, extent=[-1,1,-1,1], origin='lower')
plt.colorbar()
plt.xlabel('Y')
plt.ylabel('X')
plt.show()
```

## Quick Test Prediction

```python
test_state = torch.tensor([0.0, 0.0], dtype=torch.float32)
with torch.no_grad():
    action = model(test_state)

print('Test action:', action)
```
